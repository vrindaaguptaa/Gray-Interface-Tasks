```python
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["text_wo_punct"] = df["text_lower"].apply(lambda text: remove_punctuation(text))
df.head()
```

**Bigram Language Model for prediction of next word without using PyTorch**

```python
bgrams={}
for i in range(len(words)-2):
  w1 = words[i]
  w2 = words[i+1]

  if not w1 in bgrams:
    bgrams[w1] = {}

  if not w2 in bgrams[w1]:
    bgrams[w1][w2] = 1
  else:
    bgrams[w1][w2] +=1
```

```python
print(bgrams['book'])
```

```python
for w in bgrams:
  bgrams[w] = dict(sorted(bgrams[w].items(), key=lambda item: -item[1]))
  total = sum(bgrams[w].values())
  bgrams[w] = dict([(k, bgrams[w][k]/total) for k in bgrams[w]])
```

```python
bgrams
```

```python
import random

def next_word(word):
  vars = bgrams[word]
  return random.choices(list(vars.keys()), weights=vars.values())[0]
```

**Bigram Language Model Using PyTorch for prediction of next word**

```python
import torch
import torch.nn as nn
import torch.optim as optim
from collections import defaultdict
import random
```

```python
word_to_idx = {word: i for i, word in enumerate(set(words))}
idx_to_word = {i: word for word, i in word_to_idx.items()}
```

```python
word_indices = [word_to_idx[word] for word in words]
```

```python
bigram_pairs = [(word_indices[i], word_indices[i+1]) for i in range(len(word_indices)-1)]
```

```python
class BigramLanguageModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim):
        super(BigramLanguageModel, self).__init__()
        # Embedding layer to represent words in a dense vector space
        self.embeddings = nn.Embedding(vocab_size, embedding_dim)
        # Linear layer for predicting the next word
        self.linear = nn.Linear(embedding_dim, vocab_size)

    def forward(self, inputs):
        embeds = self.embeddings(inputs)
        out = self.linear(embeds)
        return out
```

```python
vocab_size = len(word_to_idx)
embedding_dim = 10  # Embedding size
model = BigramLanguageModel(vocab_size, embedding_dim)
```

```python
loss_function = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
inputs = torch.tensor([pair[0] for pair in bigram_pairs], dtype=torch.long)
targets = torch.tensor([pair[1] for pair in bigram_pairs], dtype=torch.long)
```

```python
epochs = 100
for epoch in range(epochs):
    model.zero_grad()

    # Forward pass
    output = model(inputs)

    # Compute the loss
    loss = loss_function(output, targets)

    # Backward pass and optimize
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
```

```python
def predict_next_word(model, word):
    model.eval()  # Set the model to evaluation mode
    with torch.no_grad():
        word_idx = torch.tensor([word_to_idx[word]], dtype=torch.long)
        output = model(word_idx)
        predicted_idx = torch.argmax(output, dim=1).item()
        return idx_to_word[predicted_idx]
```

```python
input_word = 'book'
predicted_word = predict_next_word(model, input_word)
print(f"Next word after '{input_word}' is '{predicted_word}'")
```
