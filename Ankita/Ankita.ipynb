```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# %matplotlib inline

df = pd.read_csv('/content/archive (1).zip')

df.info()

df.drop(['Name', 'Sex', 'Embarked', 'Cabin', 'Ticket'], axis=1, inplace=True)

df['Fare'].fillna(df['Fare'].mode()[0],inplace=True)

sns.boxplot(x='Pclass', y='Age', data=df)

def impute_age(cols):
  Age = cols[0]
  Pclass= cols[1]
  if pd.isnull(Age):
    if Pclass == 1:
      return 41
    elif Pclass == 2:
      return 26
    else:
      return 23
  else:
    return Age

df['Age'] = df[['Age', 'Pclass']].apply(impute_age, axis=1)

df.info()

"""*List of parameters name from a function in python"""

import inspect

print(inspect.signature(impute_age))

"""*To print multiple arguments in python"""

def print_data(dataset):
  if isinstance(dataset, pd.DataFrame):
    for index, row in dataset.iterrows():
      print("Age:", row['Age'], "-Survived", row['Survived'])
  else:
    print('Input not in dataframe.')

print_data(df)

"""Python program to find the power of a number using recursion"""

def power(base, expo):
  if expo == 0:
    return 1
  elif expo == 1:
    return base
  else:
    return base*power(base, expo-1)

df.head(4)

for index, row in df.iterrows():
  print(power(row['Age'], row['Pclass']))

```
