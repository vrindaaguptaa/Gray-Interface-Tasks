# Alice's Adventures in Wonderland

```python
with open('/kaggle/input/dementor-image-harry-potter/Alice in the Wonderland.txt', 'r') as text_file:
    text = text_file.read()
```

```python
import re   # for regular expressions
from collections import Counter
```

```python
# first we gotta remove any unwanted characters and convert the entire text to lowercase
text = re.sub(r'[^a-zA-Z\s.,?]', '', text).lower()  # cleaning the dataset
```

```python
words = re.findall(r'\b\w+[\w.,?]\b', text)

# the above expression finds all sequences of word characters that maybe followed by word characters, periods, commas, question marks.  \b denotes the word boundaries
```

```python
words  # currently a list of all the words in the text
```

```python
# to create the bigram -- 

bigrams = [(words[i], words[i+1]) for i in range(len(words)-1)]      # list of tuple 

bigrams    # currently has a list of tuples having the adjacent words
```

```python
bigram_counts = Counter(bigrams)
```

```python
type(bigram_counts)
```

```python
bigram_counts
```

```python
print(bigram_counts.most_common(10))
```

```python
from collections import defaultdict
```

```python
bigram_freq = defaultdict(Counter)
```

```python
for (w1, w2), count in bigram_counts.items():
    bigram_freq[w1][w2] = count
```

```python
bigram_freq

# has a word and its correponding dictionary of next words
```

```python
# converting counts to probabilites 

bigram_prob = {}
for (w1, next_words) in bigram_freq.items():
    total_count = sum(next_words.values())    # sum the total word counts for the denominator of the probability
    bigram_prob[w1] = {w2: count / total_count for w2, count in next_words.items()}    # find the probability for each word occuring after the current word
```

```python
import torch, random
```

Now that we're done with all this -- let's generate the text using this model

```python
def generate_text(start_word, bigram_prob, max_words = 200, threshold = 0.001):
    current_word = start_word
    generated_text = [current_word]
    
    for _ in range(max_words):
        if current_word not in bigram_prob:
            break
        # find the possible next words for the current word
        next_words = list(bigram_prob[current_word].keys())
        # find the probabilities for each of the next words
        probabilities = list(bigram_prob[current_word].values())
        
        # making them into a tensor
        prob_tensor = torch.tensor(probabilities)
        
        # sampling the next word --- multinomial distribution
        next_word_idx = torch.multinomial(prob_tensor, 1).item()
        next_word = next_words[next_word_idx]
        
        # checking for below threshold probabilities
        if bigram_prob[current_word][next_word] < threshold:
            break
        
        generated_text.append(next_word)
        current_word = next_word
    
    return ' '.join(generated_text)
```

```python
start_word = 'alice'
generated_text = generate_text(start_word, bigram_prob)

print(generated_text)
```
