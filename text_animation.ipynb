**GENERATING IMAGES USING PIPELINE**

```python
import PIL
img=[]
text_prompt="night sky with stars and moon"
for j in range(24):
  img.append(pipeline(text_prompt, num_inference_steps=10, num_images_per_prompt=5, output_type="pil").images)
print(len(img))
```

```python
images=[]
# Convert PIL Images to NumPy arrays
for i in range(len(img)):
  for j in range(5):
    images.append(img[i][j])
```

**CONVERTING IMAGE TO ARRAY AND CREATING SEQUENCE FROM GENERATED IMAGES**

```python
from PIL import Image
from moviepy.editor import ImageSequenceClip, concatenate_videoclips

# # List of image paths
# image_list = [
#     Image.open(images[0]),
#     Image.open(image[1]),
#     Image.open(image[2]),
#     Image.open(image[3]),
#     Image.open(image[4])
# ]

import numpy as np

image_a = [np.array(img) for img in images]
print(len(image_a))
print(image_a[0])
# Create a list of ImageClip objects from the image paths
image_clips = [ImageSequenceClip([path], fps=24) for path in image_a]

# Concatenate the image clips into a single clip
final_clip = concatenate_videoclips(image_clips)

# Set the duration of the final clip (adjust as needed)
final_clip = final_clip.set_duration(5)  # 5 seconds
print (final_clip)
# Write the final clip to a video file
#final_clip.write_videofile("animation.mp4")
```

**CONVERTING FRAME TO VIDEO**

```python

try:
    final_clip.write_videofile("./Output/animation_night_sky.mp4")
    print("Animation saved successfully as animation.mp4")
except Exception as e:
    print(f"Error occurred while saving the animation: {e}")
```
