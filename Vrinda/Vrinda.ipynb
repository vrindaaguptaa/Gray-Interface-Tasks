most repeated word in a text file.

```python

with open("/content/drive/MyDrive/Colab Notebooks/file1.txt","r") as file1:
   words=file1.read().split() # split will split the txt. file into words.
dict={} # created dictionary where words will be stored.here key:value is word:no. of times it occured.
for a in words:
  if a in dict:
    dict[a]=dict[a]+1
  else:
    dict[a]=1
max_time_word=None
max_count=0
for a, count in dict.items():
  if count>max_count:
    max_count=count
    max_time_word=a
print("the most repeated word is", max_time_word)
print("it appears",max_count,"times")




```

To find power of a number using recursion

```python
def fun(a,b):
  if b==0:
    return 1
  else:
    return a*fun (a,b-1)
a=5
b=6
result=fun(a,b)
print("result is:", result)
```

```python
import inspect
def fun(number1,number2):
  return number1+number2
sig=inspect.signature(fun)
print(sig.parameters)
```
